## Data Wrangling

### Gathering Data

```python
customers_df = pd.read_csv("https://raw.githubusercontent.com/hakansyukur23/Project-Analisis-Data-E-Commerce/main/DataSet/customers_dataset.csv")
customers_df.head()
```

```python
order_items_df = pd.read_csv("https://raw.githubusercontent.com/hakansyukur23/Project-Analisis-Data-E-Commerce/main/DataSet/order_items_dataset.csv")
order_items_df.head()
```

```python
#Open File Order Payment
order_payment_df = pd.read_csv("https://raw.githubusercontent.com/hakansyukur23/Project-Analisis-Data-E-Commerce/main/DataSet/order_payments_dataset.csv")
order_payment_df.head()
```

```python
#Open File Order Review Dataset
order_reviews_df = pd.read_csv("https://raw.githubusercontent.com/hakansyukur23/Project-Analisis-Data-E-Commerce/main/DataSet/order_reviews_dataset.csv")
order_reviews_df.head()
```

```python
#Open Products category name translation
products_translation_df = pd.read_csv("https://raw.githubusercontent.com/hakansyukur23/Project-Analisis-Data-E-Commerce/main/DataSet/product_category_name_translation.csv")
products_translation_df.head()
```

```python
#Open Products Dataset
products_df = pd.read_csv("https://raw.githubusercontent.com/hakansyukur23/Project-Analisis-Data-E-Commerce/main/DataSet/products_dataset.csv")
products_df.head()
```

```python
#Open File Sellers Dataset
sellers_df = pd.read_csv("https://raw.githubusercontent.com/hakansyukur23/Project-Analisis-Data-E-Commerce/main/DataSet/sellers_dataset.csv")
sellers_df.head()
```

```python
#Open orders Dataset
orders_df = pd.read_csv("https://raw.githubusercontent.com/hakansyukur23/Project-Analisis-Data-E-Commerce/main/DataSet/orders_dataset.csv")
orders_df.head()
```

#### marger data

```python
gabung1_df = pd.merge(
    left=products_translation_df,
    right=products_df,
    how="left",
    left_on="product_category_name",
    right_on="product_category_name"
)
gabung1_df.drop(columns=['product_category_name', 'product_name_lenght', 'product_description_lenght', 'product_photos_qty', 'product_weight_g',	'product_length_cm',	'product_height_cm',	'product_width_cm'], inplace=True)
gabung1_df.head()
```

```python
gabung2_df = pd.merge(
    left=gabung1_df,
    right=order_items_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
gabung2_df.head()
```

```python
gabung3_df = pd.merge(
    left=gabung2_df,
    right=order_payment_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
gabung3_df.head()
```

```python
gabung4_df = pd.merge(
    left=gabung3_df,
    right=orders_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
gabung4_df.drop(columns=['order_purchase_timestamp',	'order_approved_at',	'order_delivered_carrier_date',	'order_delivered_customer_date',	'order_estimated_delivery_date'], inplace=True)
gabung4_df.head()
```

```python
gabung5_df = pd.merge(
    left=gabung4_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
gabung5_df.drop(columns=['customer_unique_id', 'customer_zip_code_prefix'], inplace=True)
gabung5_df.head()
```

```python
gabung6_df = pd.merge(
    left=gabung5_df,
    right=order_reviews_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
gabung6_df.drop(columns=['review_id',	'review_creation_date',	'review_answer_timestamp', 'review_comment_message'], inplace=True)
gabung6_df.head()
```

```python
analyst_data_df = pd.merge(
    left=gabung6_df,
    right=sellers_df,
    how="left",
    left_on="seller_id",
    right_on="seller_id"
)
analyst_data_df.drop(columns=['seller_zip_code_prefix'], inplace=True)
analyst_data_df.head()
```
